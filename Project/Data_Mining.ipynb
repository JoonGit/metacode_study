```python
import pandas as pd
import numpy as np
import re

def reduce_and_evaluate_food_data(food_df):
    # ---------------------------------------------------------
    # 1. 데이터 전처리 (컬럼 추출 및 형변환)
    # ---------------------------------------------------------
    
    # 분석용 핵심 컬럼 추출 (원본 보호를 위해 copy)
    core_columns = [
        '식품코드', '업체명', '식품명', '대표식품명', '식품대분류명', '식품중분류명', '영양성분함량기준량',
        '에너지(kcal)', '단백질(g)', '지방(g)', '탄수화물(g)', 
        '당류(g)', '식이섬유(g)', '나트륨(mg)', '칼륨(mg)', '콜레스테롤(mg)', 
        '포화지방산(g)', '트랜스지방산(g)',  '데이터기준일자'
    ]
    df_core = food_df[core_columns].copy()
    
    # 기준량 데이터 정제 (문자열에서 숫자만 추출 후 변환)
    df_core['영양성분함량기준량'] = df_core['영양성분함량기준량'].astype(str).str.extract(r'(\d+\.?\d*)')
    df_core['영양성분함량기준량'] = pd.to_numeric(df_core['영양성분함량기준량'], errors='coerce')
    
    # 영양성분 컬럼 숫자형 변환
    numeric_cols = [
        '에너지(kcal)', '단백질(g)', '지방(g)', '탄수화물(g)', 
        '당류(g)', '식이섬유(g)', '나트륨(mg)', '칼륨(mg)', '콜레스테롤(mg)', 
        '포화지방산(g)', '트랜스지방산(g)', 
    ]
    for col in numeric_cols:
        df_core[col] = pd.to_numeric(df_core[col], errors='coerce')
    
    # 결측치 0 대체 및 컬럼명 직관화
    df_core.fillna(0, inplace=True)
    df_core = df_core.rename(columns={'영양성분함량기준량': '기준량(g)'})
    
    # ---------------------------------------------------------
    # 2. 영양학적 지표 계산 (분모가 0일 때의 오류 방지를 위해 np.where 사용)
    # ---------------------------------------------------------
    
    # [2-1] 에너지 밀도 (1g당 열량)
    df_core['에너지밀도(kcal/g)'] = np.where(
        df_core['기준량(g)'] > 0, 
        df_core['에너지(kcal)'] / df_core['기준량(g)'], 
        0
    )
    
    # [2-2] 3대 영양소 에너지 비율 (%) (탄수화물 4kcal, 단백질 4kcal, 지방 9kcal 기준)
    df_core['탄수화물_에너지비율(%)'] = np.where(df_core['에너지(kcal)'] > 0, (df_core['탄수화물(g)'] * 4 / df_core['에너지(kcal)']) * 100, 0)
    df_core['단백질_에너지비율(%)'] = np.where(df_core['에너지(kcal)'] > 0, (df_core['단백질(g)'] * 4 / df_core['에너지(kcal)']) * 100, 0)
    df_core['지방_에너지비율(%)'] = np.where(df_core['에너지(kcal)'] > 0, (df_core['지방(g)'] * 9 / df_core['에너지(kcal)']) * 100, 0)
    
    # [2-3] 세부 영양 품질 지표
    # 단백질 INQ (하루 권장량: 에너지 2,000kcal, 단백질 55g 기준)
    # 1.0 이상이면 단백질이 충분히 포함된 식품으로 간주
    df_core['단백질_INQ'] = np.where(
        df_core['에너지(kcal)'] > 0,
        (df_core['단백질(g)'] / 55) / (df_core['에너지(kcal)'] / 2000),
        0
    )
    
    # 당류 비율 (전체 탄수화물 중 단순당 비중)
    df_core['당류비율(%)'] = np.where(
        df_core['탄수화물(g)'] > 0,
        (df_core['당류(g)'] / df_core['탄수화물(g)']) * 100,
        0
    )
    
    # 나트륨-칼륨 비율 (수분 저류 예측 지표, 1.0에 가까울수록 균형)
    df_core['나트륨_칼륨_비율'] = np.where(
        df_core['칼륨(mg)'] > 0,
        df_core['나트륨(mg)'] / df_core['칼륨(mg)'],
        0
    )
    
    # ---------------------------------------------------------
    # 3. 데이터 후처리
    # ---------------------------------------------------------
    
    # 무한대 값(inf) 0 치환 및 소수점 둘째 자리 반올림
    df_core.replace([np.inf, -np.inf], 0, inplace=True)
    
    return df_core.round(2)

# 실행
food_cleaned_df = reduce_and_evaluate_food_data(food_df)
food_cleaned_df.head()
```

```python
import pandas as pd

def reduce_and_clean_shop_data(shop_df):
    # ---------------------------------------------------------
    # 1. 필요 데이터 추출 (원본 보호를 위해 copy)
    # ---------------------------------------------------------
    core_columns = [
        '인허가번호', '업소명', '영업상태코드', '도로명주소', 
        '소재지주소', '전화번호', '음식의유형', '주된음식종류'
    ]
    df_core = shop_df[core_columns].copy()
    
    # ---------------------------------------------------------
    # 2. 영업상태 필터링 (1, 2: 영업 중)
    # ---------------------------------------------------------
    # 타입 혼재(숫자/문자) 오류를 방지하기 위해 문자열 변환 후 필터링
    df_core = df_core[df_core['영업상태코드'].astype(str).isin(['1', '2'])]
    
    # ---------------------------------------------------------
    # 3. 결측치 대체 및 텍스트 정제
    # ---------------------------------------------------------
    # 텍스트형 결측치(NaN)를 분석 목적에 맞게 기본값으로 치환
    fill_dict = {
        '도로명주소': '',
        '소재지주소': '',
        '전화번호': '',
        '음식의유형': '',
        '주된음식종류': ''
    }
    df_core.fillna(fill_dict, inplace=True)
    
    # 문자열 컬럼의 불필요한 양옆 공백 제거
    text_cols = df_core.select_dtypes(include=['object']).columns
    for col in text_cols:
        df_core[col] = df_core[col].astype(str).str.strip()

    # 전화번호 컬럼에서 판다스 실수형 변환으로 인해 생긴 '.0' 꼬리표 제거
    # 정규표현식 r'\.0$'는 '문자열의 끝에 있는 .0'만 타겟팅하여 빈 문자열로 바꿉니다.
    df_core['전화번호'] = df_core['전화번호'].str.replace(r'\.0$', '', regex=True)
        
    return df_core

# 실행
processed_shop_df = reduce_and_clean_shop_data(shop_df)
processed_shop_df.head()
```

```python
import pandas as pd

def reduce_restaurant_payment_data(purchase_df):
    # 1. 음식점 결제 도메인에 매핑할 원본 컬럼과 새 컬럼명 딕셔너리 정의
    # 키(Key): 원본 컬럼명, 값(Value): 변경할 음식점 결제 컬럼명
    rename_mapping = {
        '결제순번(STTLM_SEQ)': 'TID',
        '시설ID(FCLT_ID)': '식당ID', 
        '결제상품금액(AMT)': '총주문금액',
        '결제금액(STTLM_AMT)': '최종결제금액',
        '결제수단(PAYMETHOD)': '결제수단',
        '결제상태코드(STTLM_STCD)': '결제상태코드', 
        '취소금액(CANCELAMT)': '취소금액',
        '부분취소구분(PARTIALCANCELCODE)': '부분취소여부',
        '취소결과코드(CANCELRESULTCODE)': '취소상태',
        '취소사유(CANCELMSG)': '취소사유',
        '등록일시2(RGSTN_DTM)': '최초등록일시',
        '결제완료일시(STTLM_CMPLE_DTM)': '결제완료일시'
    }
    
    available_columns = [col for col in rename_mapping.keys() if col in purchase_df.columns]
    df_core = purchase_df[available_columns].copy()
    df_core = df_core.rename(columns=rename_mapping)
    
    # 4. 결제 관련 텍스트 데이터의 결측치 및 공백 정제
    # 취소 사유 등의 결측치를 빈 문자열로 처리
    fill_dict = {
        '취소금액': 0,
        '총주문금액': 0,
        '최종결제금액': 0,
        '취소사유': '',
        '부분취소여부': 'N',
        '취소상태': '',
        '결제수단': '정보없음'
    }
    # 존재하는 컬럼에 대해서만 결측치 처리
    for col, fill_val in fill_dict.items():
        if col in df_core.columns:
            df_core[col] = df_core[col].fillna(fill_val)
                
    return df_core

# 실행 예시
restaurant_payment_df = reduce_restaurant_payment_data(perchase_df)
restaurant_payment_df.head()
```

```python
import pandas as pd
import numpy as np

def build_relational_mock_db(food_df, shop_df, payment_df):
    """
    서로 연관 없는 3개의 데이터셋을 RDB 형태의(상품-상점-결제) 
    """
    # 1. 원본 훼손 방지를 위한 복사
    db_food = food_df.copy()
    db_shop = shop_df.copy()
    db_payment = payment_df.copy()
    
    # 2. 기준키(PK) 목록 추출
    # 식당 PK: 인허가번호 / 음식 PK: 식품코드
    valid_shop_ids = db_shop['인허가번호'].unique()
    valid_food_ids = db_food['식품코드'].unique()
    
    # 3. 결제 테이블(db_payment)에 인위적인 외래키(FK) 덮어쓰기 및 추가
    np.random.seed(42) # 결과 재현성을 위해 난수 시드 고정
    
    # [상점-결제 연결] 기존 식당ID를 무시하고 우리가 가진 식당의 '인허가번호'로 덮어쓰기
    db_payment['식당ID'] = np.random.choice(valid_shop_ids, size=len(db_payment))
    
    # [상품-결제 연결] 어떤 음식을 결제했는지 알 수 있도록 '식품코드' 컬럼 추가 (1결제 = 1상품 가정)
    db_payment['식품코드'] = np.random.choice(valid_food_ids, size=len(db_payment))
    
    # 4. 결제 테이블 컬럼 순서 직관적으로 재배치 (TID, 식당ID, 식품코드 순)
    cols = db_payment.columns.tolist()
    # 식당ID와 식품코드를 맨 앞으로 당겨옴
    cols.insert(1, cols.pop(cols.index('식당ID')))
    cols.insert(2, cols.pop(cols.index('식품코드')))
    db_payment = db_payment[cols]
    
    return db_food, db_shop, db_payment

# ---------------------------------------------------------
# 실행 및 검증 (Verification)
# ---------------------------------------------------------
# 앞서 만든 food_cleaned_df, processed_shop_df, restaurant_payment_df 를 넣고 실행합니다.
final_food_db, final_shop_db, final_payment_db = build_relational_mock_db(
    food_cleaned_df, 
    processed_shop_df, 
    restaurant_payment_df
)

# 1. [상점-결제 연결 검증]
is_shop_linked = final_payment_db['식당ID'].isin(final_shop_db['인허가번호']).all()
print(f"상점-결제 연결 완료 여부: {is_shop_linked}")

# 2. [상품-결제 연결 검증]
is_food_linked = final_payment_db['식품코드'].isin(final_food_db['식품코드']).all()
print(f"상품-결제 연결 완료 여부: {is_food_linked}")
```

```python
import pandas as pd
import re

def translate_columns_to_english(food_db, shop_db, payment_db):
    # 1. 영문 매핑 딕셔너리 정의
    food_rename_dict = {
        '식품코드': 'food_id', '업체명': 'company_name', '식품명': 'food_name', 
        '대표식품명': 'rep_food_name', '식품대분류명': 'major_category', 
        '식품중분류명': 'sub_category', '기준량(g)': 'serving_size_g', 
        '에너지(kcal)': 'energy_kcal', '단백질(g)': 'protein_g', '지방(g)': 'fat_g', 
        '탄수화물(g)': 'carbs_g', '당류(g)': 'sugar_g', '식이섬유(g)': 'dietary_fiber_g', 
        '나트륨(mg)': 'sodium_mg', '칼륨(mg)': 'potassium_mg', '콜레스테롤(mg)': 'cholesterol_mg', 
        '포화지방산(g)': 'sat_fat_g', '트랜스지방산(g)': 'trans_fat_g', '데이터기준일자': 'data_ref_date', 
        '에너지밀도(kcal/g)': 'energy_density_kcal_g', '탄수화물_에너지비율(%)': 'carbs_energy_ratio_pct', 
        '단백질_에너지비율(%)': 'protein_energy_ratio_pct', '지방_에너지비율(%)': 'fat_energy_ratio_pct', 
        '단백질_INQ': 'protein_inq', '당류비율(%)': 'sugar_ratio_pct', '나트륨_칼륨_비율': 'na_k_ratio'
    }
    
    shop_rename_dict = {
        '인허가번호': 'shop_id', '업소명': 'shop_name', '영업상태코드': 'biz_status_code', 
        '도로명주소': 'road_address', '소재지주소': 'land_address', 
        '전화번호': 'phone_number', '음식의유형': 'food_type', '주된음식종류': 'main_dish_type'
    }
    
    payment_rename_dict = {
        'TID': 'tid', '식당ID': 'shop_id', '식품코드': 'food_id', 
        '총주문금액': 'total_order_amt', '최종결제금액': 'final_pay_amt', 
        '결제수단': 'pay_method', '결제상태코드': 'pay_status_code', 
        '취소금액': 'cancel_amt', '부분취소여부': 'is_partial_cancel', 
        '취소상태': 'cancel_status', '취소사유': 'cancel_reason', 
        '최초등록일시': 'created_at', '결제완료일시': 'paid_at'
    }
    
    # 2. 컬럼명 변경 적용 (원본 보호를 위해 copy)
    en_food_db = food_db.copy().rename(columns=food_rename_dict)
    en_shop_db = shop_db.copy().rename(columns=shop_rename_dict)
    en_payment_db = payment_db.copy().rename(columns=payment_rename_dict)
    
    return en_food_db, en_shop_db, en_payment_db

# ---------------------------------------------------------
# 실행 및 검증 (Verification)
# ---------------------------------------------------------
# 함수 실행
en_food_db, en_shop_db, en_payment_db = translate_columns_to_english(
    final_food_db, 
    final_shop_db, 
    final_payment_db
)

# [컬럼명 영어 변환 검증] 정규표현식(영문자, 숫자, 언더스코어로만 구성) 사용
def verify_english_columns(df, db_name):
    pattern = re.compile(r'^[a-zA-Z0-9_]+$')
    all_english = all(pattern.match(col) for col in df.columns)
    print(f"{db_name} 영문 컬럼명 변환 완료 여부: {all_english}")

verify_english_columns(en_food_db, "상품(Food) 테이블")
verify_english_columns(en_shop_db, "상점(Shop) 테이블")
verify_english_columns(en_payment_db, "결제(Payment) 테이블")

# 최종 결과 확인
en_payment_db.head()
```

```python
import os
import pandas as pd
import time
from langchain_core.documents import Document
from langchain_openai import OpenAIEmbeddings
from langchain_chroma import Chroma

# ---------------------------------------------------------
# 0. 환경 변수 및 사전 데이터 준비
# ---------------------------------------------------------
print("1. 데이터 프레임 병합 및 RAG용 자연어 컨텍스트 생성 시작...")

# 1. 업체명이 있는 데이터만 필터링하여 RAG 도큐먼트 생성에 활용 (업체명 없는 데이터는 RAG 컨텍스트로 활용하기 어려움)
company_food_db = en_food_db[en_food_db['company_name'] != '해당없음']
sampled_df = company_food_db.copy()

# 2. [page_content 정적 데이터 최적화 & 메타데이터 분리]
def build_general_rag_documents(df):
    docs = []
    for _, row in df.iterrows():
        page_content = f"""
브랜드: {row['company_name']}
메뉴 특징: {row['major_category']} 분류에 속하는 {row['food_name']} 메뉴입니다.
영양 및 열량: 1회 제공량 {row['serving_size_g']}g 기준으로 {row['energy_kcal']}kcal의 에너지를 냅니다.
주요 성분: 단백질 {row['protein_g']}g, 탄수화물 {row['carbs_g']}g, 지방 {row['fat_g']}g이 포함되어 균형 잡힌 식사가 가능합니다.
""".strip()
        metadata = {
            "food_id": str(row['food_id']),
            "company_name": str(row['company_name']),
            "food_name": str(row['food_name']),
            "protein_inq": float(row['protein_inq']),
            "na_k_ratio": float(row['na_k_ratio'])   
        }
        
        docs.append(Document(page_content=page_content, metadata=metadata))
    return docs

# 문서 변환 실행
documents = build_general_rag_documents(sampled_df)
print(f"-> 총 {len(documents)}개의 범용 RAG 도큐먼트 변환 완료 (업체명 있는 데이터 적용).")

# 3. 임베딩 및 Chroma DB 구축
embeddings = OpenAIEmbeddings(model="text-embedding-3-small")

DB_PATH = "./food_db"

# [수정된 부분] 한 번에 적재하던 from_documents를 제거하고, 빈 DB 생성 후 분할 적재
db = Chroma(
    embedding_function=embeddings,
    persist_directory=DB_PATH,
    collection_metadata={"hnsw:space": "cosine"},
    collection_name="food_db_V2" 
)

print("-> Vector DB 분할 적재 시작 (TPM 제한 방지)...")
batch_size = 300  # 한 번에 처리할 도큐먼트 개수 (한도 초과 시 이 숫자를 줄이세요)

for i in range(0, len(documents), batch_size):
    batch_docs = documents[i : i + batch_size]
    db.add_documents(batch_docs)
    print(f"   - {min(i + batch_size, len(documents))} / {len(documents)} 개 적재 완료. (잠시 대기 중...)")
    time.sleep(2)  # 2초 대기하여 1분당 토큰 발생량을 분산

print("-> Vector DB 전체 적재 완료.\n")

# ---------------------------------------------------------
# [추가됨] 과제 제출용 정량적 검증 (Quantitative Verification)
# ---------------------------------------------------------
print("="*50)
print(" [정량적 검증] Vector DB 구축 결과 확인")
print("="*50)

# 1) 임베딩 차원(Dimension) 검증
sample_embed = embeddings.embed_query("테스트")
model_dim = len(sample_embed)

db_sample = db._collection.get(limit=1, include=['embeddings'])

if db_sample.get('embeddings') is not None and len(db_sample['embeddings']) > 0:
    db_dim = len(db_sample['embeddings'][0])
else:
    db_dim = 0

print(f"▶ 임베딩 모델({embeddings.model}) 기본 차원 수: {model_dim}")
print(f"▶ Vector DB 적재 차원 수: {db_dim}")
print(f"☞ 차원 일치 여부: {'일치 (PASS)' if model_dim == db_dim else '불일치 (FAIL)'}")
print("-" * 50)

# 2) 데이터 개수(Count) 검증
total_chunks = len(documents)
db_count = db._collection.count()

print(f"▶ 텍스트 스플리터/생성된 총 청크 수: {total_chunks}개")
print(f"▶ Vector DB 최종 적재 도큐먼트 수: {db_count}개")
print(f"☞ 개수 일치 여부: {'일치 (PASS)' if total_chunks == db_count else '불일치 (FAIL)'}")
print("="*50 + "\n")

# ---------------------------------------------------------
# 4. 범용 유사도 검색 테스트 (Qualitative Verification)
# ---------------------------------------------------------
print("="*50)
print(" [정성적 검증] 유사도 검색(Similarity Search) 테스트")
print("="*50)

test_query = "투썸플레이스에 커피음료 추천해줘"
print(f"Q: '{test_query}'\n")

# k=3 개 추출하여 점수 확인
results = db.similarity_search_with_score(test_query, k=3)

for i, (doc, distance) in enumerate(results, 1):
    # 코사인 거리(Cosine Distance)를 코사인 유사도(Cosine Similarity)로 변환
    similarity = 1 - distance
    
    print(f"[{i}순위] 메뉴: {doc.metadata['company_name']} - {doc.metadata['food_name']}")
    print(f"  - 유사도 점수: {similarity:.4f} " + ("(적합👍)" if similarity >= 0.7 else "(다소 낮음)"))
    print("-" * 40)
```

```python
import os
import pandas as pd
from langchain_core.documents import Document
from langchain_openai import OpenAIEmbeddings, ChatOpenAI
from langchain_chroma import Chroma
from langchain_classic.chains.query_constructor.base import AttributeInfo
from langchain_classic.retrievers.self_query.base import SelfQueryRetriever
from langchain_community.query_constructors.chroma import ChromaTranslator

# ---------------------------------------------------------
# 4. Self-Querying Retriever 구축 및 통합 테스트
# ---------------------------------------------------------
print("\n[Self-Querying Retriever 기반 지능형 검색 테스트]")

# 1. [메타데이터 속성 정의] LLM이 메타데이터를 어떻게 필터링할지 알 수 있도록 친절한 설명서(Schema) 작성
metadata_field_info = [
    AttributeInfo(
        name="company_name",
        description="음식점의 브랜드명 또는 업체명 (예: 투썸플레이스, 교촌치킨 등)",
        type="string",
    ),
    AttributeInfo(
        name="food_name",
        description="메뉴의 고유 이름",
        type="string",
    ),
    AttributeInfo(
        name="protein_inq",
        description="단백질 영양 품질 지표. 수치가 1.0 이상이면 고단백 및 영양이 우수함을 의미함.",
        type="float",
    ),
    AttributeInfo(
        name="na_k_ratio",
        description="나트륨-칼륨 비율. 수치가 낮을수록 건강한 식단임을 의미함.",
        type="float",
    ),
]

# 수정된 부분 1: 문서의 성격을 LLM에게 알려주는 설명으로 변경
document_content_description = "국내 음식점 및 카페 메뉴의 영양 성분, 칼로리, 그리고 메뉴에 대한 세부 설명 정보"

# 2. 쿼리 분석용 LLM 초기화 (비용과 속도가 효율적인 gpt-4o-mini 추천)
llm = ChatOpenAI(temperature=0, model="gpt-4o-mini")

# 3. [리트리버 생성] LLM, Vector DB, 메타데이터 스키마를 결합
retriever = SelfQueryRetriever.from_llm(
    llm,
    db,
    document_content_description,
    metadata_field_info,
    structured_query_translator=ChromaTranslator(), 
    enable_limit=True 
)

# 4. [자연어 필터링 통합 테스트] 
test_query = "투썸플레이스 메뉴 중 커피음료 추천해줘"
print(f"Q: '{test_query}'\n")

# 리트리버 실행 (similarity_search_with_score 대신 invoke 사용)
results = retriever.invoke(test_query)

# 결과 출력
for i, doc in enumerate(results, 1):
    print(f"[{i}순위] 메뉴: {doc.metadata['company_name']} - {doc.metadata['food_name']}")
    print("-" * 40)

# similarity_search_with_score 결과
# [1순위] 메뉴: 투썸플레이스 - 커피_카페모카 아이스(ICED) (R)
#   - 유사도 점수: 0.5858 (다소 낮음)
# ----------------------------------------
# [2순위] 메뉴: 투썸플레이스 - 커피_롱블랫 아이스(ICED) (Short)
#   - 유사도 점수: 0.5771 (다소 낮음)
# ----------------------------------------
# [3순위] 메뉴: 투썸플레이스 - 커피_카페모카 아이스(ICED) (L)
#   - 유사도 점수: 0.5769 (다소 낮음)
# ----------------------------------------
```
